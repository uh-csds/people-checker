```python
from bs4 import BeautifulSoup
import requests

info_url = "https://www.helsinki.fi/en/about-us/people/people-finder?researchOrganizations.id[]=62060775&search=social%20data%20science"

# Fetch the HTML content of the webpage
response = requests.get(info_url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(response.content, 'html.parser')
    print(soup)
    # Find all instances of the class "ah-search-results__title"
    titles = soup.find_all(class_="ah-search-results__title")
    
    # Print the titles
    for title in titles:
        print(title.text.strip())  # Use strip() to remove leading/trailing whitespaces
else:
    print("Failed to retrieve webpage.")

reserach_id = 62060775
```
